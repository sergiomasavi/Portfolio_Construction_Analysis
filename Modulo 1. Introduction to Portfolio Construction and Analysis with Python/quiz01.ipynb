__Pregunta 1. ¿Cuál fue el rendimiento anualizado de la cartera Lo 20 durante todo el período?__

Respuesta: 15.20%

```python
n_months = df_ffme.shape[0]
return_per_month = (df_ffme+1).prod()**(1/n_months) - 1
annualized_return = (return_per_month + 1)**12-1
print('Annualized returns:',annualized_return,sep=2*'\n')
```

```python
# Calcular volatilidad anualizada para comparar ambos activos
annualized_vol = df_ffme.std()*np.sqrt(12)
print('Volatilidad anualizada',annualized_vol,sep='\n')
```

__Para las preguntas restantes, utilice el conjunto de datos de los Índices de Fondos de Cobertura EDHEC.__

```python
df_hfi = erk.get_hfi_returns(directory='data/')

time1 = '2009'

df = df_hfi.loc[(df_hfi.index>=time1)].copy()
```

__Respuesta 13. Observando los datos desde 2009 (incluyendo todo el año 2009) hasta 2018, ¿qué índice de fondos de cobertura ha mostrado la mayor semidesviación?__

Respuesta: Short Selling

```python
erk.negative_semideviation(df).sort_values(ascending=False)
```

__Pregunta 14. Si observamos los datos desde 2009 (incluyendo todo el año 2009), ¿qué índice de fondos de cobertura ha mostrado la menor semidesviación?__

Respuesta: Fixed Income Arbitrage

__Pregunta 15. Si se observan los datos desde 2009 (incluyendo todo el año 2009), ¿cuál es el índice de fondos de cobertura que ha tenido un valor de asimetría más negativo?__

Respuesta: Equity Market Neutral

```python
erk.skewness(df).sort_values(ascending=False)
```

__Pregunta 16. Si se observan los datos desde el año 2000 (incluyendo todo el año 2000) hasta 2018, ¿qué índice de fondos de cobertura ha mostrado la mayor curtosis?__

Respuesta: Fixed Income Arbitrage

```python
time1 = '2000'

df = df_hfi.loc[(df_hfi.index>=time1)].copy()
erk.kurtosis(df).sort_values(ascending=False)
```
